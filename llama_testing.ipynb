```python
import pandas as pd
from llama_index.core import Document, VectorStoreIndex
import openai
import os
from dotenv import load_dotenv
load_dotenv()

# Set your OpenAI API key
openai.api_key = os.getenv("OPENAI_API_KEY")

# Example DataFrame with restaurant data

df = pd.read_csv('/Users/pulakdugar/Documents/pearhackathon/Foodie-Finder/my_dataframe.csv')
df.drop(["ID", "Phone"], axis = 1, inplace = True)
df.dropna(axis=0, inplace=True)
df['text'] = "Name: " + df['Name'] + " - Location: " + df['Location'] + " - Rating: " + df['Rating'].astype(str) + " - # of reviews: " + df['Review Count'].astype(str) + " - Price: " + df['Price'] + " - Categories: " + df['Categories'] + " - Reviews: " + df['Reviews']

df
```

```python
# Create documents from the DataFrame
documents = [Document(text=row['text']) for _, row in df.iterrows()]

# print(documents)
# Build the index
index = VectorStoreIndex.from_documents(documents)

# Query the index and generate a response
query = "Which restaurant is rated the highest in berkeley?"
eng = index.as_chat_engine()
response = eng.query(query)

# Output the response
print(response)
```
